### - Remove Ids

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['date'].head()
```

```python
df['year']=df['date'].apply(lambda date: date.year)
```

```python
df['month']=df['date'].apply(lambda date: date.month)
```

```python
df.drop('date', axis=1, inplace=True)
```

```python
df.drop('zipcode', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.shape
```

### - Split Data into Features and label

```python
X=df.drop('price', axis=1).values #features
y=df['price'].values #label
```

```python
X.shape
```

### Split data into train test sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=0.7, random_state=25)
```

### - Scale Input values

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

### Model Creation and Training(Fitting) 

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))
model.add(Dense(units=19, activation='relu'))

model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mse') ## mse is usually the preferred for regression type problems
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=500)
```

```python
loss_df=pd.DataFrame(model.history.history)
loss_df.head()
```

```python
loss_df.plot()
```
